```python
import json
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
file_path = Path (f'D:/data/vis_data/20231204_222652/vis_data/scalars.json')
```

```python
import json
import matplotlib.pyplot as plt

# File path


# Initialize lists to store the extracted data
aAcc_values = []
mIoU_values = []
mAcc_values = []
steps = []

# Read the file and parse each line
with open(file_path, 'r') as file:
    for line in file:
        data = json.loads(line)
        if 'aAcc' in data and 'mIoU' in data and 'mAcc' in data:
            aAcc_values.append(data['aAcc'])
            mIoU_values.append(data['mIoU'])
            mAcc_values.append(data['mAcc'])
            steps.append(data['step'])

# Plotting the data
# plt.plot(steps, aAcc_values, label='aAcc')
plt.plot(steps, mIoU_values, label='mIoU')
plt.plot(steps, mAcc_values, label='mAcc')

plt.xlabel('Step')
plt.ylabel('Values')
plt.title('Graph of aAcc, mIoU, and mAcc')
plt.legend()
plt.show()
```
